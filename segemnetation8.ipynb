##Seuillage simple


```python
import cv2
import numpy as np

```

```python
image=cv2.imread("omate.jpg")
```

```python
from google.colab.patches import cv2_imshow
```

```python
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)




```

```python
def seuilleg_simple(img,seuil, couleur, option):
  return cv2.threshold(img,seuil, couleur, option)
```

```python
ret, thresh1 =  seuilleg_simple(img, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 =  seuilleg_simple(img, 120, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 =  seuilleg_simple(img, 120, 255, cv2.THRESH_TRUNC)
ret, thresh4 =  seuilleg_simple(img, 120, 255, cv2.THRESH_TOZERO)
ret, thresh5 = seuilleg_simple(img, 120, 255, cv2.THRESH_TOZERO_INV)
```

```python

cv2_imshow( thresh1)
cv2_imshow( thresh2)
cv2_imshow(thresh3)
cv2_imshow( thresh4)
cv2_imshow( thresh5)
```

##Seuillage adaptatif

```python
def seuillage_adaptif( img, maxValue, adaptiveMethod, thresholdType, blockSize, C):
  return cv2.adaptiveThreshold( img, maxValue, adaptiveMethod, thresholdType, blockSize, C)
```

```python
seug1= seuillage_adaptif(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                        cv2.THRESH_BINARY, 199, 5)

seug2= seuillage_adaptif(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                        cv2.THRESH_BINARY, 199, 5)
```

```python
cv2_imshow(seug1)
cv2_imshow(seug2)
```

Dans le seuillage simple, une valeur de seuil globale constante est employe tout au long du processus. Cependant, cette approche se rvle inefficace lorsque les conditions d'clairage varient d'une zone  l'autre. C'est l qu'intervient le seuillage adaptatif, une mthode qui calcule des seuils spcifiques  des zones plus restreintes. En consquence, cela engendre des valeurs de seuil distinctes pour chaque rgion en fonction des variations d'clairage observes.
