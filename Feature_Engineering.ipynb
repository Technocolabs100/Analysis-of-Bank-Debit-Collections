```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 1: Standardize the data
scaler = StandardScaler()
standardized_data = scaler.fit_transform(df)
```

```python
# Step 2: Calculate the covariance matrix
covariance_matrix = np.cov(df, rowvar=False)
```

```python
covariance_matrix
```

```python
# Step 2: Calculate eigenvectors and eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvectors and eigenvalues in decreasing order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Now you have eigenvalues and eigenvectors, which are the core results of PCA

# Example: Print eigenvalues and first principal component
print("Eigenvalues:")
print(eigenvalues)

print("\nFirst Principal Component (Eigenvector):")
print(eigenvectors[:, 0])
```

```python
# Calculate the cumulative explained variance
total_variance = np.sum(eigenvalues)
explained_variance_ratio = eigenvalues / total_variance
cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)

# Determine the number of dimensions that explain 90% of the variance
desired_variance_explained = 0.9  # Change this as needed
num_dimensions_to_keep = np.argmax(cumulative_explained_variance_ratio >= desired_variance_explained) + 1

# Select the top-k eigenvectors to form the feature vector
feature_vector = eigenvectors[:, :num_dimensions_to_keep]

print("Feature Vector:")
print(feature_vector)
print("Number of Dimensions Kept:", num_dimensions_to_keep)
print("Cumulative Explained Variance Ratio:", cumulative_explained_variance_ratio[num_dimensions_to_keep - 1])
```

```python
# Recast the data along the principal component axes
transformed_data = np.dot(df, feature_vector)

print("Transformed Data:")
print(transformed_data)
```

```python
from sklearn.feature_selection import mutual_info_classif

#Separate the target variable and the features
X = df
y = df['class']

#Calculate Mutual Information scores
mi_scores = mutual_info_classif(X, y)

#Create a DataFrame to store the feature names and their corresponding MI scores
mi_df = pd.DataFrame({'Feature': X.columns, 'MI Score': mi_scores})

#Sort the DataFrame by MI Score in descending order
mi_df = mi_df.sort_values(by='MI Score', ascending=False)

#Print the result
print(mi_df)
```

```python
threshold = 0.005

# Select top features above the threshold
top_features = mi_df[mi_df['MI Score'] > threshold]['Feature']

# Create a subset of data using the selected top features
selected_data = df[top_features]

selected_data

```
