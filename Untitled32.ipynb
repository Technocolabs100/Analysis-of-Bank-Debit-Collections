```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Dataset_Tp_4.csv")
```

```python
df.head(20)
```

```python
categorical_columns = df.select_dtypes(include=['category', 'object']).columns
```

```python
categorical_columns
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Create a ColumnTransformer with OneHotEncoder
ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(sparse=False), categorical_columns)],
    remainder='passthrough'  # Keep non-categorical columns as is
)

# Fit and transform the data
encoded_data = ct.fit_transform(df)

# Get the feature names for the encoded categorical columns
encoder = ct.named_transformers_['encoder']
feature_names = encoder.get_feature_names_out(input_features=categorical_columns)

# Combine the feature names with the names of non-categorical columns
all_feature_names = list(feature_names) + list(df.columns.drop(categorical_columns))

# Create a new DataFrame with the encoded data and feature names
encoded_df = pd.DataFrame(encoded_data, columns=all_feature_names)
```

```python
encoded_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(encoded_df)
```

```python
len(scaled_data[0])
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Créez un objet PCA en spécifiant le nombre de composantes principales à extraire
n_components = len(scaled_data[0])  # Vous pouvez ajuster le nombre de composantes ici
pca = PCA(n_components=n_components)

# Appliquez l'ACP sur vos données
pca.fit(scaled_data)

# Obtenez le ratio de variance expliquée par chaque composante principale
explained_variance_ratio = pca.explained_variance_ratio_

# Affichez les ratios de variance expliquée
for i, ratio in enumerate(explained_variance_ratio):
    print(f"Composante Principale {i + 1}: {ratio:.4f}")

# Vous pouvez également calculer la variance cumulée expliquée
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
print("Variance cumulée expliquée :")
for i, cum_var in enumerate(cumulative_explained_variance):
    print(f"Jusqu'à la composante {i + 1}: {cum_var:.4f}")
```

```python
import matplotlib.pyplot as plt

# Utilisez le code de l'ACP précédent pour obtenir les ratios de variance expliquée
# et la variance cumulée expliquée

# Créez un graphique de la variance cumulée
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Variance Cumulée de l\'ACP')
plt.xlabel('Nombre de Composantes Principales')
plt.ylabel('Variance Cumulée Expliquée')
plt.grid(True)
plt.show()
```

### Le nombre optimal de composantes principales (dans votre cas, 17) est choisi de manière à conserver une quantité significative de la variabilité des données tout en réduisant la dimensionnalité de manière efficace. Cela vous permet de simplifier les données sans perdre d'informations importantes, ce qui est souvent utile dans l'analyse des données et la modélisation.
