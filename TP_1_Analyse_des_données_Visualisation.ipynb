# Importation des bibliothèques

Dans cette partie du code, nous allons importer les librairies nécessaires pour la visualisation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importation des données

Pour importer les données, nous utiliserons la fonction read() de pandas.

```python
data = pd.read_csv("covid19.csv")
data.head(10)
```

```python
#l'indice du maroc est 117
X = data[116:117]
X
```

```python
# select morocco's data
x = data.loc[data["Country/Region"]=='Morocco']
x
```

```python
fig, ax = plt.subplots(figsize=(15,8))
x.plot(y=["Confirmed","Deaths","Recovered","Active","New cases","New deaths", "New recovered"],kind='bar',ax=ax)
```

```python
fig, ax = plt.subplots(1, figsize=(50, 15))
sns.barplot(x = "Country/Region" , y="Recovered", data = data[:5], orient = 'v')
```

```python
fig, ax = plt.subplots(1, figsize=(50, 15))
sns.barplot(y = "Country/Region" , x="Recovered", data = data[:5], orient = 'h')
```

```python
data_sorted = data.sort_values(by=['Deaths'],ascending = False)
data_sorted
```

```python
fig, ax = plt.subplots(figsize=(30,10))
sns.barplot(x='Country/Region',y = 'Deaths',data = data_sorted[:10])
```

```python
fig, ax = plt.subplots(figsize=(30,10))
sns.barplot(y='Country/Region',x = 'Deaths',data = data_sorted[:15],orient ='h')
```

```python
import matplotlib.patches as mpatches
df_grouped = data_sorted.groupby('WHO Region').sum()[['Confirmed', 'Deaths', 'Recovered']]
df_WHORegion= df_grouped[:10]
df_WHORegion
```

```python
df_WHORegion.plot(kind = 'barh' , stacked = True)
```

```python
data_afrique = data[data['WHO Region'] == 'Africa']
data_afrique.shape
ig, ax = plt.subplots(1,3,figsize=(30,16))
ax[0].set_title('Nb Deaths')
ax[0].pie(data_afrique['Deaths'])

ax[1].set_title('Comfirmed')
ax[1].pie(data_afrique['Confirmed'])

ax[2].set_title('Recovered')
ax[2].pie(data_afrique['Recovered'])

fig.legend(data_afrique['Country/Region'])
```

```python
#fig_dims = (20, 10)
fig, ax = plt.subplots(figsize=(20, 10))
sns.violinplot(y = data["Deaths"], x = data["WHO Region"])
plt.show()
```

```python
#fig_dims = (20, 10)
fig, ax = plt.subplots(figsize=(20, 10))
sns.boxplot(y = data["Deaths"], x = data["WHO Region"])
plt.show()
```
